```python
%pylab inline
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data =pd.read_json('apps.json')    
```

```python
data_category = data['category']
count =data_category.value_counts().plot(count, kind = 'bar', title= 'Free i-Tune Apps' , color = 'g' , alpha = 0.8 , )
count.set_xlabel("Category")
count.set_ylabel("# of Free Apps on iTune Store")
```
